# A. Libraries Import

```python
# working on datasets
import pandas as pd
pd.set_option('display.max_column', None)
import numpy as np

# working on model import
import pickle
with open('model.pkl', 'rb') as file_1:
    model = pickle.load(file_1)

# working on dimensionality reduction
from sklearn.decomposition import PCA

display(model)
```

All essential libraries, as well as the model itself, are ready to be utilized within this notebook.

# B. Data Creation

```python
np.random.seed(12)
var1=np.random.choice(['Female', 'Male'],10) # Gender
var2=np.random.choice(['Loyal Customer', 'disloyal Customer'],10) # Customer Type
var3=np.random.randint(7,85,10) # Age
var4=np.random.choice(['Business travel', 'Personal Travel'],10) # Type of Travel
var5=np.random.choice(['Business', 'Eco', 'Eco Plus'],10) # Class
var6=np.random.randint(50,6951,10) # Flight Distance
var7=np.random.randint(0,5,10) # Seat comfort
var8=np.random.randint(0,5,10) # Departure/Arrival time convenient
var9=np.random.randint(0,5,10) # Food and drink
var10=np.random.randint(0,5,10) # Gate location
var11=np.random.randint(0,5,10) # Inflight wifi service
var12=np.random.randint(0,5,10) # Inflight entertainment
var13=np.random.randint(0,5,10) # Online support
var14=np.random.randint(0,5,10) # Ease of Online booking
var15=np.random.randint(0,5,10) # On-board service
var16=np.random.randint(0,5,10) # Leg room service
var17=np.random.randint(1,5,10) # Baggage handling
var18=np.random.randint(0,5,10) # Checkin service
var19=np.random.randint(0,5,10) # Cleanliness
var20=np.random.randint(0,5,10) # Online boarding
var21=np.random.randint(0,1592,10) # Departure Delay in Minutes
var22=np.random.randint(0,1584,10) # Arrival Delay in Minutes
data = {'Gender':var1,
        'Customer Type':var2,
        'Age':var3,
        'Type of Travel':var4,
        'Class':var5,
        'Flight Distance':var6,
        'Seat comfort':var7,
        'Departure/Arrival time convenient':var8,
        'Food and drink':var9,
        'Gate location':var10,
        'Inflight wifi service':var11,
        'Inflight entertainment':var12,
        'Online support':var13,
        'Ease of Online booking':var14,
        'On-board service':var15,
        'Leg room service':var16,
        'Baggage handling':var17,
        'Checkin service':var18,
        'Cleanliness':var19,
        'Online boarding':var20,
        'Departure Delay in Minutes':var21,
        'Arrival Delay in Minute':var22}
```

```python
inf_df = pd.DataFrame(data=data)
inf_df
```

We have generated our inference data using randomized imputation with numpy.

# C. Data Prediction

```python
inf_pred = model.predict(inf_df)
inf_pred
```

We have obtained predictions for our dataset in the form of an array, ready to be integrated with the main inference dataset.

# D. Prediction Evaluation

```python
# combine prediction result to the main inference dataset
final_inf_df = inf_df.copy()
new_col_pred = []
for i in range(len(inf_pred)):
    if inf_pred[i] == 1:
        new_col_pred.append('Dissatisfied')
    else:
        new_col_pred.append('Satisfied')
final_inf_df['Satisfaction'] = new_col_pred
final_inf_df
```

Our prediction result has been concatenated with the main inference dataset.
